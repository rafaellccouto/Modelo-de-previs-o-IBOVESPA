```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o conjunto de dados
df = pd.read_csv('Ibovespa.csv')

print('As 5 primeiras linhas do DataFrame:')
print(df.head())

print('\nValores ausentes em cada coluna:')
print(df.isnull().sum())

print('\nTipos de dados de cada coluna:')
df.info()

print('\nEstatísticas descritivas:')
print(df.describe())

# Converter a coluna 'Data' para objetos datetime, especificando o formato
df['Data'] = pd.to_datetime(df['Data'], format='%d.%m.%Y')

# Ordenar o DataFrame pela coluna 'Data'
df = df.sort_values(by='Data').reset_index(drop=True)

print('\nDataFrame após converter a Data para datetime e ordenar:')
print(df.head())

# Visualizar as principais tendências ao longo do tempo
plt.figure(figsize=(12, 6))
plt.plot(df['Data'], df['Último'])
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')
plt.title('Preço de Fechamento do IBOVESPA ao Longo do Tempo')
plt.grid(True)
plt.show()
```

```python
def clean_volume(volume_str):
    if isinstance(volume_str, str):
        volume_str = volume_str.replace(',', '.')
        if 'B' in volume_str:
            return float(volume_str.replace('B', '')) * 1_000_000_000
        elif 'M' in volume_str:
            return float(volume_str.replace('M', '')) * 1_000_000
        else:
            try:
                return float(volume_str)
            except ValueError:
                return None # Lidar com casos onde a conversão falha
    return volume_str # Retornar como está se não for uma string

# Aplicar a função à coluna 'Vol.'
df['Vol.'] = df['Vol.'].apply(clean_volume)

# Converter a coluna 'Var%' para float numérico
df['Var%'] = df['Var%'].str.replace(',', '.').str.replace('%', '').astype(float)

# Calcular a variação diária no preço 'Último'
daily_change = df['Último'].diff()

# Criar a variável alvo 'Trend' (1 para aumento, 0 para diminuição/sem mudança)
df['Trend'] = (daily_change > 0).astype(int)

# Criar 'Daily_Range' (Amplitude Diária)
df['Daily_Range'] = df['Máxima'] - df['Mínima']

# Criar 'Open_Close_Diff' (Diferença Abertura-Fechamento)
df['Open_Close_Diff'] = df['Último'] - df['Abertura']

# Calcular as Médias Móveis Simples (SMA)
df['SMA_5'] = df['Último'].rolling(window=5).mean()
df['SMA_10'] = df['Último'].rolling(window=10).mean()
df['SMA_20'] = df['Último'].rolling(window=20).mean()

# Criar atributo defasado para o preço 'Último'
df['Último_Lag1'] = df['Último'].shift(1)

# Remover linhas com valores NaN resultantes da criação de atributos
df.dropna(inplace=True)

print('DataFrame após engenharia de atributos e remoção de NaNs:')
print(df.head())
print('\nTipos de dados após engenharia de atributos:')
df.info()
print('\nEstatísticas descritivas após engenharia de atributos:')
print(df.describe())
```

```python
from sklearn.preprocessing import StandardScaler

# 1. Definir atributos (X) e alvo (y)
X = df.drop(['Data', 'Último', 'Trend'], axis=1)
y = df['Trend']

# 2. Dividir dados em conjuntos de treinamento e teste (últimos 30 dias para teste)
test_size = 30

X_train = X.iloc[:-test_size]
X_test = X.iloc[-test_size:]
y_train = y.iloc[:-test_size]
y_test = y.iloc[-test_size:]

print(f"Formato de X_train: {X_train.shape}")
print(f"Formato de X_test: {X_test.shape}")
print(f"Formato de y_train: {y_train.shape}")
print(f"Formato de y_test: {y_test.shape}")

# 3. Normalizar atributos numéricos
scaler = StandardScaler()

# Ajustar nos dados de treinamento e transformar tanto os dados de treinamento quanto os de teste
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Converter arrays normalizados de volta para DataFrames, preservando nomes de colunas e índice
X_train = pd.DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
X_test = pd.DataFrame(X_test_scaled, columns=X_test.columns, index=X_test.index)

print('\nPrimeiras 5 linhas de X_train após a normalização:')
print(X_train.head())
print('\nPrimeiras 5 linhas de X_test após a normalização:')
print(X_test.head())
```

```python
from sklearn.ensemble import RandomForestClassifier

# Instanciar um modelo RandomForestClassifier
# Usando alguns parâmetros comuns para um bom ponto de partida
model = RandomForestClassifier(n_estimators=100, random_state=42, max_depth=10)

print(f"Modelo Selecionado: {model}")
```

O RandomForestClassifier foi escolhido por sua robustez e capacidade de lidar com relações complexas e não lineares frequentemente encontradas em dados financeiros de séries temporais. Ele é menos propenso a overfitting em comparação com árvores de decisão únicas e pode fornecer insights sobre a importância dos atributos, o que é valioso para entender os drivers do mercado. Sua natureza de conjunto também ajuda a fazer previsões mais generalizadas e estáveis para tarefas de classificação, como a previsão de tendências.

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Treinar o modelo
model.fit(X_train, y_train)

# 2. Fazer previsões nos dados de teste
y_pred = model.predict(X_test)

# 3. Calcular e imprimir a acurácia
accuracy = accuracy_score(y_test, y_pred)
print(f"Acurácia: {accuracy:.2f}")

# 4. Calcular e imprimir o relatório de classificação
print("\nRelatório de Classificação:")
print(classification_report(y_test, y_pred))

# 5. Gerar e exibir uma matriz de confusão
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Diminuir/Sem Mudança', 'Aumentar'], yticklabels=['Diminuir/Sem Mudança', 'Aumentar'])
plt.xlabel('Previsto')
plt.ylabel('Real')
plt.title('Matriz de Confusão')
plt.show()
```

A natureza sequencial dos dados foi abordada de várias maneiras durante a engenharia de atributos e a divisão dos dados:

  Engenharia de Atributos: Para capturar as dependências temporais inerentes, vários atributos específicos de séries temporais foram criados:

  Atributos Defasados (Lagged Features): O atributo Último_Lag1, que representa o preço 'Último' (fechamento) do dia anterior, incorpora diretamente informações passadas na observação atual, permitindo que o modelo compreenda as mudanças diárias.
   Médias Móveis: Médias Móveis Simples (SMAs) em diferentes janelas (SMA_5, SMA_10, SMA_20) foram calculadas. Esses atributos suavizam as flutuações de preço e destacam tendências, fornecendo ao modelo insights sobre o momentum de curto e médio prazo do IBOVESPA.

Divisão Temporal dos Dados: Crucialmente, os dados foram divididos em conjuntos de treinamento e teste reservando os últimos 30 dias para teste. Isso respeita estritamente a ordem cronológica dos dados da série temporal. Essa abordagem é vital para evitar vazamento de dados, garantindo que o modelo seja avaliado em dados futuros não vistos, o que é uma simulação realista de seu desempenho em um cenário de negociação ao vivo.

Ao incorporar esses atributos sequenciais, um modelo não específico para séries temporais, como o RandomForestClassifier, pode aprender efetivamente os padrões temporais nos dados. Os preços defasados e as médias móveis atuam como proxies para a dimensão do tempo, fornecendo ao modelo o contexto histórico necessário para fazer previsões informadas sobre a tendência futura, apesar de não ser um modelo inerentemente sequencial. O modelo aprende as relações entre esses indicadores históricos e a tendência futura, imitando como um modelo de série temporal usaria observações passadas.


### Compromissos entre Acurácia e Overfitting

1.  **Overfitting em Previsões de Séries Temporais e Implicações de 100% de Acurácia:**
    Em previsões de séries temporais, o overfitting ocorre quando um modelo aprende demais os dados de treino, incluindo ruído e flutuações aleatórias, resultando em ótimo desempenho no conjunto de treino mas fraca generalização para dados não vistos. Alcançar 100% de acurácia no conjunto de teste para uma tarefa de previsão de tendência, como a do IBOVESPA, é altamente incomum e frequentemente indica uma forte possibilidade de overfitting ou vazamento de dados (data leakage). Mercados financeiros são inerentemente ruidosos e não determinísticos, tornando a previsão perfeita quase impossível. Uma acurácia tão alta sugere que o modelo pode ter aprendido inadvertidamente padrões específicos do contexto histórico do conjunto de teste, que podem não se manter para dados futuros verdadeiramente não vistos. É crucial ser cético em relação a pontuações perfeitas em cenários do mundo real.

2.  **Trade-offs entre Acurácia e Overfitting:**
    O principal trade-off é entre a complexidade do modelo e sua capacidade de generalização. Um modelo muito complexo (por exemplo, uma rede neural profunda com muitas camadas, ou uma Random Forest com altos valores de `n_estimators` e `max_depth`) pode atingir acurácia muito elevada nos dados de treino ao capturar padrões intrincados. Porém, se essa complexidade levar o modelo a aprender ruído em vez das tendências subjacentes, ocorrerá overfitting. Um modelo mais simples pode apresentar menor acurácia no treino, mas generalizar melhor para novos dados. O objetivo é encontrar um equilíbrio: ser complexo o suficiente para capturar padrões relevantes, mas não ao ponto de memorizar os dados de treino. Para dados financeiros, uma acurácia ligeiramente menor e mais realista, que generalize bem, é muito mais valiosa do que uma pontuação perfeita porém overfitted.

3.  **Medidas Já Tomadas para Mitigar Overfitting:**
    *   **Divisão Temporal:** Ao reservar os últimos 30 dias de dados para teste (`test_size = 30`), garantimos uma divisão temporal, o que é crucial para séries temporais. Isso previne vazamento de informação do futuro para o conjunto de treino, imitando um cenário real de previsão. Se uma divisão aleatória fosse usada, informações futuras poderiam influenciar sutilmente o treino, resultando em desempenho excessivamente otimista.
    *   **Parâmetros do RandomForestClassifier:** A escolha do `RandomForestClassifier` oferece alguma resistência inerente ao overfitting em comparação com árvores de decisão únicas, já que agrega previsões de múltiplas árvores. Os parâmetros `n_estimators=100` (número de árvores) e `max_depth=10` (profundidade máxima de cada árvore) foram explicitamente definidos. `max_depth` limita o quão profundas as árvores podem crescer, evitando que se tornem excessivamente específicas aos dados de treino. Embora `n_estimators` possa aumentar a complexidade do modelo, um valor razoável como 100 ajuda a diversificar o ensemble sem necessariamente causar overfitting severo, especialmente quando as árvores individuais são limitadas por `max_depth`.

4.  **Medidas Adicionais para Iterações Futuras:**
    Dada a acurácia de 100%, recomenda-se fortemente investigação adicional e estratégias de mitigação:
    *   **Validação Cruzada para Séries Temporais:** Em vez de uma única divisão treino-teste, empregar técnicas de validação cruzada específicas para séries temporais (por exemplo, rolling origin ou walk-forward validation) fornecerá uma estimativa mais robusta da capacidade de generalização do modelo. Isso envolve re-treinar o modelo iterativamente em janelas de dados crescentes e testar em períodos subsequentes não vistos.
    *   **Ajuste de Hiperparâmetros com Validação Cruzada:** Ajustar sistematicamente hiperparâmetros (por exemplo, `max_features`, `min_samples_leaf`, `max_depth`, `n_estimators`) do `RandomForestClassifier` usando validação cruzada para séries temporais (por exemplo, `GridSearchCV` ou `RandomizedSearchCV` com um splitter temporal personalizado) ajudará a encontrar o equilíbrio ideal entre viés e variância. Isso pode revelar se os parâmetros atuais estão realmente levando a um modelo overfitted.
    *   **Modelos Mais Simples:** Se o desempenho do modelo permanecer suspeitosamente alto após validações adicionais, considerar modelos mais simples como Regressão Logística ou um modelo baseado em árvore mais simples pode ser benéfico. Às vezes, um modelo mais simples que capture as tendências principais generaliza melhor e é mais interpretável do que um modelo excessivamente complexo, especialmente em domínios altamente voláteis como mercados financeiros.
    *   **Análise de Importância de Atributos:** Embora a Random Forest forneça importância de atributos, uma análise mais profunda para garantir que nenhum atributo espúrio ou altamente correlacionado esteja influenciando excessivamente as previsões pode ser útil. Reduzir o número de atributos ou aplicar técnicas de redução de dimensionalidade também pode ajudar.
    *   **Regularização:** Embora Random Forests não possuam parâmetros de regularização diretos como L1/L2, controlar `max_depth`, `min_samples_leaf` e `min_samples_split` tem efeito similar ao limitar a complexidade das árvores. Experimentar com esses valores pode ajudar a prevenir overfitting.
